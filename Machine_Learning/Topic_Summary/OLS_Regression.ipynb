# **OLS(ordinary least square) Regression의 공식**

- 최소자승법(OLS, Ordinary Least Squares) 이란, 산점도를 통해 데이터의 분포 그래프를 그릴때, 이 데이터들의 경향을 알기 위한 최적의 추세선을 그리기 위한 방법 중 하나.
- OLS는 근사적으로 구하려는 해와 실제 해의 오차의 제곱의 합이 최소가 되는 해를 구하는 방법

- OLS Regression은 회귀를 통해서 방정식의 상수 값들을 추정하는 데에 사용.
- n개의 입력값과 그에 대응하는 출력값 $(x_{i},y_{i})(1\leq i\leq n)$이 있고, 이 계의 방정식이 변수 $x$와 $\beta=(\beta_{0},\beta_{1},\cdots ,\beta_{k})$인 상수 $\beta$에 대한 식 $f(x, \beta)$으로 주어질 때, $\sum {i}(y{i}-f(x_{i}, \beta))^{2}$ 의 값을 최소로 만드는 $\beta$를 구하는 것이 문제의 목표.

**OLS vs. MSE**

- OLS(Ordinary Least Square): 선형 회귀 모델을 만들기 위한 선횡 최소 제곱법, 모델을 만들때 사용.
- MSE(Mean Square Error): 모델 성능 평가 지표, 모델을 평가할 때 사용.
