### **Helm 이란?**

* helm이란 쿠버네티스 패키지 매니저.
* centOS에서 yum 이나 python에서는 pip 툴 MAC OS에서는 brew와 같이 패키지를 관리 주는 툴.
* helm을 이용하면 원하는 패키지들을 쿠버네티스에 쉽게 설치할 수 있다.

* helm은 docker hub와 비슷하게 helm 패키지들을 저장하고 있는 저장소(repository)가 있다.
* 사용자는 저장소를 추가하고 해당 저장소의 패키지를 install하기만 하면 됨.
* helm 차트로 원하는 패키지를 install할때 values.yaml 을 이용하여 사용자의 환경에 따라 커스텀하여 사용할 수 있다.

### Helm 폴더 구조

```yaml
sample main ✗ ➤ tree .
.
├── Chart.yaml
├── charts
├── templates
│   ├── NOTES.txt
│   ├── _helpers.tpl
│   ├── deployment.yaml
│   ├── hpa.yaml
│   ├── ingress.yaml
│   ├── service.yaml
│   ├── serviceaccount.yaml
│   └── tests
│       └── test-connection.yaml
└── values.yaml

3 directories, 10 files
```

**1. charts (dir)**

* charts는 sample (가칭) 이라는 charts에서 사용하는 sub Charts 등을 모아두는 공간이 된다.

* 우리가 만약 어떠한 서비스 제품을 install 할 때 필요한 third party 부분의 chart라고  생각하면 편하다.

* a라는 서비스를 Deploy 하기 위해서는 mariadb가 필요하다.

* mariadb를 따로 서비스 Deploy를 할 수 도 있지만 관리를 2번 해야한다는 번거로움이 생김으로 a라는 서비스를 Deploy시 mariadb를 같이 Deploy 시키고 싶다.

* 이런경우 a라는 chart 안에 charts 라는 디렉토리에 mariadb charts를 넣어둔다면 한번의 install 명령어로 설치가 가능 해진다.

**2. templates (dir)**

* Helm Chart는 뼈대와 살이라는 표현을 사용했다.

* templates에서는 주로 k8s 리소스등의 뼈대를 구성하여 작성한다.

**3. values.yaml**

* 뼈대에 살을 구성하기위한 value들의 모음집이다.

* 각 프로젝트 성격에 따라 복수개로 관리할수도 있다.

**4. Chart.yaml**

* Chart에 대한 기본정보를 담고있는 Yaml 파일이다.

* chart의 이름, chart의 버전, application의 버전, 의존성 등등을 작성해논다.

### HELM 사용해보기

```yaml
helm create <chart명> # helm chart 생성

helm install <chart명> <helm chart 경로>
```


GCP에 helm chart 배포하기

```yaml
helm package sample
export HELM_EXPERIMENTAL_OCI=1
gcloud auth print-access-token | helm registry login -u oauth2accesstoken \                   
--password-stdin https://asia-northeast3-docker.pkg.dev
helm push sample-0.1.0.tgz oci://asia-northeast3-docker.pkg.dev/rokit-pjt-labs-2022-08-01/helm
helm install sample oci://asia-northeast3-docker.pkg.dev/rokit-pjt-labs-2022-08-01/helm/sample --version 0.1.0
```
