### **Kustomize란**

kustomize는 yaml에서 사용하는 **manifest를 관리 하는 도구를 말한다.**

### Kustomize 구성

다음과 같이 폴더 및 파일을 생성

```bash
nginx-kustomize main ✗ ➤ tree .
.
├── base
│   ├── deployment.yaml
│   ├── kustomization.yaml
│   └── service.yaml
└── overlays
    ├── dev
    │   └── kustomization.yaml
    └── prod
        └── kustomization.yaml

4 directories, 5 files
```

**쿠버네티스 리소스 생성해보기**

kustomize로 쿠버네티스 리소스를 생성하는 명령어.

```bash
kubectl kustomzie ./
```

위 명령어를 사용하면 다음 예시와 유사한 것들을 볼 수 있다.


image.png


쿠버네티스에 적용을 원할 경우 다음과 같이 작성.

```bash
kubectl kustomize ./ | kubectl apply -f -
```

**쿠버네티스 리소스 삭제해보기**

kustomize 실행결과를 쉘 파이프라인과 kubectl delete -로 쿠버네티스 리소스를 삭제.

```bash
kubectl kustomize ./ | kubectl delete -f -
```

### **kustomization.yaml파일** 작성하기

kustomization.yaml에는 여러가지 필드를 사용할 수 있다. 공식 문서에는 기본 필드를는 4개

4개의 기본 필드

- resources 필드: kustomize를 적용할 쿠버네티스 리소스(yaml파일)
- generators 필드: 새로 생성할 필드를 설정
- transformers 필드: 기존 필드 변경을 설정
- validators 필드: 검증

필드가 진행되는 순서는 **`resources → generators → transformers → validators`**로 진행.
