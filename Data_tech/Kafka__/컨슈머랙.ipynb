# **컨슈머 랙이란?**

운영 모니터링 지표

파티션에 데이터가 하나씩 들어가게 되면 각 데이터에 오프셋이라는 숫자가 붙는데_파티션이 한개인 토픽에 프로듀서가 데이터를 넣을경우 0부터 시작

image.png

프로듀서 데이터 삽입 속도 < 컨슈머 데이터 가저가는 속도?
1. 프로듀서가 넣은 데이터의 오프셋
2. 컨슈머가 가져간 데이터의 오프셋

1_2사이의 차이가 발생 = 컨슈머 랙 (consumer lag)

랙의 숫자를 통해 해당 토픽에 대한 프로듀서와 컨슈머의 상태 유추 가능

토픽에 여러 파티션이 존재할 경우 랙이 여러개 존재 할 수 있음_한개의 토픽과 컨슈머 그룹에 대한 lag이 여러개 존재할 때, 그 중 높은 숫자의 lag을 records-lag-max라고함

컨슈머가 성능이 안나오거나 비정상 동작을 하게되면 lag이 발생한 경우가 많음음
