# **여러 네트워크 topology**

컴퓨터끼리 정보를 교환하고 교류하는 형태를 의미하는 네트워크에서의 토폴로지는 컴퓨터들의 특정한 망구성 방식을 의미.

하나의 네트워크 구성 방식을 보더라도, 노드와 링크같은 물리적 배치로구분하는 물리적 토폴로지와 노드간 데이터 흐름으로 구분하는 논리적 토폴로지로 네트워크 구성을 각각 판단할 수 있다.

**Star**: 중앙에 위치한 메인 노드를 통해 다른노드와 소통할 수 있는 구조

* 장점: 장애 발견이 쉽고 관리가 용이
* 단점: 메인 노드에 장애가 발생하면 전체 네트워크 사용 불가능

image.png

**Bus**: 버스라는 공통 배선을 통해 노드들이 연결되어 있어, 한 노드의 신호가 모든 노드에 전달.(타켓 노드만 신호에 반응하고 다른노드들은 무시)

* 장점: 노드 추가 및 삭제가 용이, 한 노드에 장애가 발생해도 다른 노드에 영향을 주지 않음.
* 단점: 공통 배선의 크기(대역폭)가 제한되어 있어, 배선에 과부하가 걸릴 경우 네트워크 성능 저하

image.png

**Ring**: 각 노드가 양 옆으로 연결된 원형 구조, 단방향으로 신호가 전달됨.

* 장점: 단방향 구조로 단순하고, 중간에 있는 노드들이 증폭기의 역할을 함(거리 제약이 적어짐)
* 단점: 노드 추가 및 삭제가 어려움

image.png

**Mesh**: 다수의 노드가 서로 연결된 형태(모두 연결 - 완전연결형, 일부 연결 - 부분 연결형)

* 장점: 노드의 장애애 영향받지않으며 유연한 대처가 가능하고 안정적임
* 단점: 구축 비용이 크고, 노드 추가에도 비용이 많이듦
