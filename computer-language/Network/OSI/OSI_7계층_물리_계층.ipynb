- 물리 계층은 OSI 7계층 중 가장 아래.
- 시스템 간의 물리적 연결을 하고, 전기 신호의 변환 및 제어 역할을 담당.
- 데이터를 '전기신호'로 바꾸어 전송하는 일을 함.

**두대의 컴퓨터가 통신할 수 있는 조건은?**

- 네트워크 통신에는 0과 1만 사용되며, 잘 주고받을 수 있으면 된다. 0과 1은 흔히 비트(bit) 라고 불림.
- 물리 계층은 이러한 0과 1이 다른 컴퓨터로 전달되는데에 핵심적인 역할을 함.
- 0과 1은 신호를 통해 주고받을 수 있는데, 신호의 종류는 아날로그 신호 와 디지털 신호가 있음.

# **디지털 신호**

image.png

- 정확히 0 혹은 1의 값을 갖음.

**디지털 신호와 현실의 괴리**

- 0과 1만 보낼 수 있으면, 두 대의 컴퓨터가 통신할 수 있고, 모든 정보를 보낼 수 있음.
- 그렇기 때문에 통신을 위해 디지털 신호를 쓸 것.
- 디지털 신호를 보내기 위해서 먼저 디지털 신호로 특정한 의미를 갖는 0과 1의 조합을 만들어 봄.
 - 예를들어, 3이라는 데이터를 보내고 싶다면 11, 8이라는 데이터를 보내고 싶다면, 100 과 같은 조합을 만들어서 보낼 수 있음.
 - 디지털 신호의 0과 1을 이용해 의미를 부여. 그런데 어떻게 보내야 할까?

**디지털 신호를 전송하는 방법 (전기 신호)**

- 전기 신호를 흘림
 - 네트워크를 구축하고 싶은 두 장치에 전기가 통할 수 있는 전선을 연결.
 - 그리고 전기 신호를 흘리면 1, 전기 신호를 흘리지 않으면 0의 통신 규약.

- 그런데 전기 신호가 흘렀는지 판단하는 기준
 - 전기라는 것은 정확히 흐르고 안 흐르고를 정의하기 어려움.
 - 현실의 전압은 1V, 2V 딱 떨어지지 않음. 수많은 소숫점이 존재.
 - 대략 5V라는 기준 전압을 세우고, 이 전압을 넘어서면, 전기가 흘렀다.
 - 이 전압을 넘어서지 못하면, 전기가 흐르지 않았다. 라는 기준을 세움
   - 이러한 기준 전압에 대해 간혹 문턱 전압이라는 표현도 씀.

**해석을 위해 주파수가 필요**
- 5V의 전기 신호가 얼마만큼의 시간동안 흘러야 디지털 신호의 1이라는 값이 될까?
  - 0.1초만 유지되면 1로 해석해야 하는가?
  - 1초를 온전히 흘러야 1로 해석해야 하는가?
  - 0.1초만 5V 밑으로 유지되면 0으로 해석해야 하는가?
  - 1초를 온전히 5V 밑으로 유지해야 0으로 해석해야 하는가?

- 주파수 개념이 쓰임.



image.png

- 예를들어, 1초간 2번의 진동을 할 수 있는 파동(2Hz의 주파수)이라는 것을 알면 해당 파동을 해석할 수 있음.
- 위의 경우 11이라는 디지털 신호가 들어왔다고 가정.

**현실에서의 주파수의 한계**
- 이론상 0~무한대 까지의 주파수를 갖는 디지털 신호를 마음대로 만들어낼 수 있지만, 현실 세계에 있는 매개체들은 그 신호를 전부 감당할 수 없음.
  - 예를들어, 인간은 약 20~20000Hz의 가청 주파수를 갖고 있는데, 그 이상의 주파수가 유입되거나 그 이하의 주파수는 유입되어도 전혀 인지하지 못함.
  - 컴퓨터의 네트워크도 무한한 주파수의 디지털 신호를 한번에 보낼 수 없고, 전선이 감당할 수 있는 적당한 단위로 끊어서 보내야 함.

**전선 말고 케이블**


> 케이블: 전기, 전자 전기가 통하지 아니하는 물질로 겉을 감싼 전화선이나 전력선


- 실제로 물리 계층에서 네트워크를 연결할 때는 케이블을 이용.
 - 주로 흔히 볼 수 있는 UTP 케이블을 이용.

- UTP 케이블은 Cat3 부터 Cat7까지 존재하는데, 보통 카테고리3, 카테고리7 등으로 발음.
 - 일반적으로 쓰는 UTP 케이블의 종류는 카테고리5.
 - 카테고리5 UTP 케이블은 최대 속도가 100Mbps이다, 주파수로 환산하면 100Mhz.
- bps 단위는 bit per second
- B/s 단위는 byte per second.

- 네트워크에 사용되는 케이블은 UTP 말고도 동축 케이블, 광케이블, STP 케이블 등 다양.



**UTP 케이블은 보통 4개의 선만 사용.**
- UTP 케이블은 총 8개의 구리선을 갖고 있지만, 실제로는 4개만 사용.
 - 일반적인 경우는 1,2,3,6 번 구리선만 각각 송수신을 위해 이용.

**UTP 케이블은 크로스와 다이렉트 케이블로 나뉨.**
- 크로스 케이블은 PC간의 직접 연결에 쓰이며, 다이렉트 케이블은 스위치와의 연결에 쓰임.
- 다이렉트 케이블은 구리선 8개를 같은 순서로 연결한 케이블이며, 크로스 케이블은 1, 2번에 연결되는 구리선을 다른쪽 커넥터의 3, 6번에 연결한 케이블.
- PC간의 연결에 크로스 케이블을 사용하는 이유는, PC에서는 동일하게 각각 1,2번 구리선을 수신에 이용하고 3,6번 구리선을 송신에 이용.
 - 이 경우 다이렉트 케이블을 직접 연결하여 이용하면 두 PC 모두 1,2번을 수신으로 쓰고, 3,6번을 송신으로 쓰기 때문에 제대로 송수신이 되지 않음.
 - 1,2번 구리선을 3,6번에 연결하고, 3,6번 구리선을 1,2번에 연결하면 올바르게 송수신이 맞물리게 됨.

- 스위치는 PC와 다르게 3,6 번을 송신에 사용하므로, 다이렉트로 만들어도 송수신이 됨.

**랜카드**

- 컴퓨터가 0과 1을 이용해 송신할 데이터를 만들면 디지털 신호로 된 데이터를 전송하기 위해 전기를 흘려야 함.
- 전기는 전선을 타고 갈 것이며, 전선은 UTP 케이블을 말함.
- 컴퓨터 본체를 살펴보면 UTP 케이블이 꽂혀있는 곳이 보임.
- UTP 케이블이 꽂혀있는 곳인 랜카드가 디지털 신호를 전기 신호로 만들어주는 주체.

- 랜카드는 보내야 하는 디지털 신호를 아날로그 전기 신호로 바꾸고, 바꾼 전기 신호를 UTP 케이블을 통해 날린다.

# **허브와 리피터**

**리피터의 역할**

- 전선이 매우 길면 전기는 전선을 타고가며 점점 손실이 일어남.
- 이러한 손실에 의해 전송에 문제가 생기는 것을 방지하기 위해 리피터는 세기가 약해진 전기 신호를 받아서 다시 강화시켜주는 역할.
- 약해진 파형을 다시 정상으로 만들어줌.
 - 요즘에는 리피터가 따로 나오는 것이 아니라, 다른 네트워크 장비에서 리피터의 역할까지 맡는 경우가 많음.

**허브의 역할**

- 컴퓨터를 여러대 연결할 수 있는 네트워크 장비.
- 허브는 리피터처럼 전기신호를 정형하고 증폭하는 기능도 함.
- 1:1로 데이터를 전송하는 것이 아니라 여러 컴퓨터의 집합을 만들어서 서로 데이터를 공유할 때 유용.
- 허브는 내가 보내고 싶은 목적지 컴퓨터에만 신호를 보내는 것이 아니라, 허브에 연결된 모든 컴퓨터에 신호를 보내서 더미 허브(dummy hub) 라는 이름으로 불리기도 함.
- 이러한 단점을 보완한 스위치가 있음.



# **물리 계층 정리**
- 두 대의 컴퓨터가 통신하기 위해서는 0과 1을 서로 주고 받을 수 있으면 됨.
- 컴퓨터는 송신을 위한 0과 1로 된 디지털 신호를 갖고 있음.
- 0과 1로 된 디지털 신호를 보내기 위해서는 전선을 통해 전달해야 함.
- 전달을 위해 전기를 이용한 아날로그 신호로 바꾸어주어야 함.
- 여기서 사용되는 네트워크 장비가 랜카드.
- 랜카드는 디지털 신호를 변환하고 전선에 해당하는 UTP 케이블로 아날로그 신호를 전달.
- 0과 1에 해당하는 아날로그 신호를 보냈으면 일단 물리계층에서 할 일은 끝남.
