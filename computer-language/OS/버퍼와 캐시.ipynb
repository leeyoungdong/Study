# **캐시와 버퍼의 차이점**
* 캐시와 버퍼는 모두 임시 저장 영역이지만 여러 면에서 다름.

**Buffer**
* 버퍼는 주로 램에서 발견되며 CPU가 일시적으로 데이터를 저장할 수 있는 영역 역할
*  예를 들어 컴퓨터와 다른 장치의 속도가 다를 때 주로 다른 출력 장치에 대한 데이터를 의미. 이렇게 하면 컴퓨터가 다른 작업을 수행할 수 있다.

* 일시적으로 데이터 저장
* 속도가 다른 장치 사이
* 입/출력 프로세스 에 사용
* 버퍼는 램의 일부.[예: 키보드 입력]

**Cache**
* 캐시는 주 메모리 또는 하드 디스크와 같은 다른 별도의 저장 영역의 일부가 될 수 있는 고속 저장 영역입니다. 
* 두 가지 캐싱 방법을 각각 메모리 캐싱 및 디스크 캐싱이라고 합니다.

* 고속 저장 영역
* 두 가지 방법
 * 메인 메모리 영역: 메모리 캐싱
 * 저장 영역: 디스크 캐싱
* 디스크에서 프로세스를 읽고 쓰는 동안 사용

**요약**
* 캐시는 고속 저장 영역 이고 버퍼는 임시 저장을 위한 RAM의 일반 저장 영역 .
* 캐시는 버퍼에 사용되는 느린 동적 램보다 빠른 정적 램으로 만들어짐.
* 버퍼는 주로 입/출력 프로세스에 사용되는 반면 캐시는 디스크에서 프로세스를 읽고 쓰는 동안 사용.
* 캐시는 디스크의 한 부분일 수도 있고 버퍼는 램의 한 부분일 수도 있음.
* 키보드에서 버퍼를 사용하여 입력 실수를 편집할 수 있지만 캐시는 편집할 수 없음.
