# **lambda function**

* 익명 함수(이름이 없는 함수)를 람다 함수라고 함.

* 람다 함수는 def 키워드를 통해서 함수를 생성하는 리터럴 표기법을 딱 한 줄의 코드로 표현할 수 있게 해주며, lambda 인자 : 표현식의 형식으로 표현.

* 람다함수는 결과 부분을 return 키워드 없이 자동으로 return. 람다함수를 사용하면 코드가 간결해지고 메모리가 절약된다는 장점이 존재.

* 그러나 함수에 이름이 없고, 저장된 변수가 없기 때문에 다시 사용하기 위해서는 다시 코드를 적어주거나, 람다함수를 변수에 담아주어야함.

*  재사용할 이유가 없다면 lambda 함수를 생성하여 넘겨주는 편이 좋음음.

image.png

```python
a = lambda x, y : x + y
print(a(5, 6))
```

# **self in Python**

```python
class MyClass:
    def __init__(self, make):
        self.make = make

    def method(make):
        return 'instance method', make

obj = MyClass
print(obj.method(1))
```

* self는 인스턴스 메서드(instance method)의 첫 번째 인자이다. 메서드가 호출될 때, 파이썬은 self에 인스턴스를 넣고 이 인스턴스를 참조하여 인스턴스 메서드를 실행할 수 있게 된다.
