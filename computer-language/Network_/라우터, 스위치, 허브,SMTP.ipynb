# **라우터와 스위치, 허브의 차이는?**

여기서의 네트워크는 LAN(Local Area Network)

image.png

**라우터(Router)**는 network layer 3계층 장비로, 네트워크 사이를 연결하는 장비. 최종 도착지의 네트워크에 도착할 수 있도록 적절한 경로를 설정하여 패킷을 전송.

**스위치(Switch)**는 data link layer 2계층 장비로, 네트워크 내에서 패킷을 전송하는 장비. 스위치로 요청이 들어오면 IP주소에 대응하는 MAC 주소를 찾아 해당 MAC 주소로 패킷을 전송한다. 만약 IP주소에 대응되는 MAC주소가 없다면 허브처럼 브로드 캐스트 방식으로 패킷을 전송하고 IP주소와 MAC주소를 대응 시킨 테이블을 갱신시킨다.

**허브(Hub)**는 physical layer 1계층 장비로 여러 기기를 연결하여 네트워크를 만들어주는 장치. 패킷을 받으면 연결된 모든 기기에 패킷을 전송함. 

브로드캐스트(Broadcast) - LAN에 있는 모든 네트워크 장비들에게 보내는 통신

# **SMTP란?**

Simple Mail Transfer Protocol은 인터넷에서 이메일을 보내기 위해 사용하는 TCP/IP프로토콜이다. TCP Port번호는 25번이다.

명렁어

* MAIL: 주소반환 확립
* RCPT: 메시지 수신자 확립
* DATA: 메시지 텍스트의 첫 신호 제공
