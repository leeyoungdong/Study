# **AWS 관리형 서비스 공급자(MSP)**



AWS 관리형 서비스 공급자(MSP) 프로그램은 입증된 레코드와 경험으로 AWS 파트너를 검증하여 계획 및 설계, 구축 및 마이그레이션, 운영 및 지원, 자동화 및 최적화를 포함하여 클라우드 여정의 모든 단계에서 고객에게 엔드 투 엔드 AWS 솔루션을 제공한다.



# **AMI의 launchPermission 속성**

각 AMI에는 소유자를 제외하고 해당 AMI를 사용한 인스턴스 시작이 허용된 launchPermission 계정을 설정할 수 있는 AWS 속성이 존재한다.

AMI의 launchPermission 속성을 변경하여 이 AMI를 퍼블릭 설정(모든 AWS 계정에 시작 권한 허용)하거나 사용자가 지정한 AWS 계정과만 공유할 수 있다.



# **​AWS Systems Manager OpsCenter**



AWS 고객이 여러 서비스에 걸친 문제, 이벤트 및 경보 집계 서비스.

고객은 단일 위치에서 문제를 검토, 조사 및 해결할 수 있으므로 서로 다른 여러 AWS 서비스로 이동할 필요가 감소된다.

관리 콘솔에서 문제, 이벤트 및 경보 등의 정보는 OpsItem(운영 항목)으로 표시되며 컨텍스트 정보, 과거 기록에 기반한 안내 및 빠른 해결 단계를

제공하며, 이 기능은 단일 위치에서 주요 조사 데이터를 제공함으로써, 해결에 소요되는 평균 시간을 단축하고, 엔지니어의 생산성을 향상하도록 설계되었다.




# **Saving Plans**

Savings Plans은 예약 인스턴스보다 훨씬 할인이 포괄적으로 적용된다.

Saving Planse은 시간당 사용 금액을 미리 약정하고, 이 약정한 금액에 대해서는 리전, 인스턴스 세대, 사이즈에 상관없이 요금 할인이 적용된다.

# **보안 그룹**

인스턴스에 대한 인바운드 및 아웃바운드 트래픽을 제어하는 가상 방화벽 역할.

 VPC에서 인스턴스를 시작할 때 최대 5개의 보안 그룹에 인스턴스를 할당할 수 있다.
