# **DynamoDB 특정 시간 복구 (PITR)**

특정 시점으로 복구를 사용하면 우발적인 쓰기 또는 삭제 작업으로부터 DynamoDB 테이블을 보호 가능.

- Amazon DynamoDB 테이블에 대한 온디맨드 백업을 생성하거나 특정 시점으로 복구를 사용한 연속 백업을 활성화.

- 특정 시점으로 복구를 설정해 두면 온디맨드 백업의 생성, 유지 관리, 예약을 걱정할 필요가 없다.

- 특정 시점 작업은 성능이나 API 지연 시간에 영향을 주지 않는다.



keyword - 데이터 손상의 경우 RPO(복구 시점 목표), RTO(복구 시간 목표)



# **Amazon DynamoDB 온디맨드**

Amazon DynamoDB 온디맨드는 용량 계획 없이 초당 수천 개의 요청을 처리할 수 있는 유연한 결제 옵션이다.

DynamoDB 온디맨드는 읽기 및 쓰기 요청에 대해 요청당 지불 요금을 제공하므로 사용한 만큼만 비용을 지불하면 된다.

온디맨드 모드를 선택하면 DynamoDB는 워크로드가 이전에 도달한 트래픽 수준으로 증가하거나 감소할 때 즉시

수용한다. 워크로드의 트래픽 수준이 새로운 피크에 도달하면 DynamoDB는 워크로드를 수용하도록 빠르게 적응한다.


**On-demand mode (주문형 모드)가 적합한 환경**

- 알 수 없는 작업 부하가 있는 새 테이블을 만듭니다.

- 예측할 수 없는 애플리케이션 트래픽이 있습니다.

- 사용한 만큼만 지불하는 간편한 방식을 선호합니다.

# **Amazon S3 File Gateway**



Amazon S3 파일 게이트웨이는 Amazon S3 에 대한 파일 인터페이스를 지원하고 서비스와 가상 소프트웨어 어플라이언스를 결합한다.

이 조합을 사용하면 NFS(네트워크 파일 시스템) 및 SMB(서버 메시지 블록)와 같은 업계 표준 파일 프로토콜을 사용하여 Amazon S3에서 객체를

저장하고 검색할 수 있다.

- Amazon S3의 파일 스토리지를 간소화하고 업계 표준 파일 시스템 프로토콜을 통해 기존 애플리케이션에 통합하며 온프레미스  스토리지에 대한 비용 효율적인 대안을 제공

- 투명한 로컬 캐싱을 통해 데이터에 대한 대기 시간이 짧은 액세스를 제공

- AWS와의 데이터 전송을 관리하고, 네트워크 정체로부터 애플리케이션을 버퍼링하고, 데이터를 병렬로 최적화 및 스트리밍하고, 대역폭 소비를 관리

keyword - SMB, NFS

**인터넷게이트웨이**

수평 확장되고 가용성이 높은 중복 VPC 구성 요소로, VPC와 인터넷 간에 통신할 수 있게 해준다. IPv4 트래픽 및 IPv6 트래픽을 지원한다.

네트워크 트래픽에 가용성 위험이나 대역폭 제약이 발생하지 않는다.

**NAT 게이트웨이**

NAT 게이트웨이는 NAT(네트워크 주소 변환) 서비스. 프라이빗 서브넷의 인스턴스가 VPC 외부의 서비스에 연결할 수 있지만

외부 서비스에서 이러한 인스턴스와의 연결을 시작할 수 없도록 NAT 게이트웨이를 사용할 수 있다.

image.png

**가상 프라이빗 게이트웨이 (Virtual Private Gateway)**

VGW는 가상 사설 네트워크인 VPC와 온프레미스 네트워크를 연결하는 방식 중 가장 오래된 방식이다.

VGW는 VPC와 온프레미스 네트워크를 연결하는 VPC 쪽의 라우터 역할을 한다.

VGW를 이용해 온프레스 네트워크와 연결하기 위해서는 온프레미스 네트워크 쪽에 CGW(Customer Gateway)를 구성해 줘야 한다.

<br>

VGW의 특징

- GW는 여러 개의 온프레미스 라우터와 연결이 가능합니다.

- 다만 VGW는 하나의 VPC에만 연결할 수 있습니다.

- VGW는 Site-to-Site VPN, Direct Connect 연결을 통해 온프레미스 네트워크와 연결이 가능합니다.

- VGW 생성시 동적 라우팅을 사용할 경우 ASN(Autonomous System Number)를 지정해 줘야합니다.



# **VPC 게이트웨이 엔드포인트**

VPC용 인터넷 게이트웨이 또는 NAT 디바이스가 없어도 Amazon S3 및 DynamoDB에 대한 안정적인 연결을 제공한다.

게이트웨이 엔드포인트는 AWS PrivateLink를 활성화하지 않는다. 게이트웨이 엔드포인트 사용에 따르는 추가 요금은 없으며

데이터 전송 및 리소스 사용량에 대한 표준 요금이 그대로 적용된다.  엔드포인트는 동일한 리전에서만 지원되며 VPC와 다른 리전의

서비스 간에 엔드포인트를 생성할 수 없습니다.

keyword - 인터넷을 통한 이동, 엑세스X



**S3 게이트웨이 엔드포인트**

VPC에서 Amazon S3 리소스로의 액세스를 이미 설정한 경우,

엔드포인트를 설정한 후에도 Amazon S3 DNS 이름을 사용하여 이러한 리소스에 액세스할 수 있다.





**Amazon S3 버킷 정책**

버킷 정책을 사용하여 특정 엔드포인트, VPC, IP 주소 범위 또는 AWS 계정의 버킷에 대한 액세스를 제어할 수 있다.



IAM등 자격 증명과 같은 인증을 사용하지 않고 S3 버킷에 비공개로 연결하도록 Amazon VPC를 구성하는 방법

: Amazon VPC에서 버킷에 액세스 하는 경우 인증하지 않고 비공개로 S3버킷에 액세스 할 수 있다.

하지만 사용된 VPC 엔드포인트가 Amazon S3를 가르키는지 확인해야 한다.



- Amazon S3에 대한 VPC 엔드포인트를 생성

- VPC 엔드포인트에서 액세스를 허용하는 버킷 정책을 추가



# **AWS PrivateLink**

지원되는 AWS 서비스, 기타 AWS 계정에서 호스팅된 서비스(VPC 엔드포인트 서비스) 및 지원되는 AWS Marketplace 파트너 서비스에 VPC를 비공개로 연결할 수 있도록 하는 가용성과 확장성이 높은 기술이다.


- 데이터를 인터넷에 노출하지 않고 VPC(Virtual Private Cloud)와 AWS 또는 온프레미스에서  호스팅되는 서비스 간에 프라이빗 연결을 설정할 수 있다.

- VPC 엔드포인트를 사용하면 인터넷 게이트웨이, NAT 디바이스, VPN 연결, AWS Direct Connect 연결하지 않고 VPC를 비공개적으로 연결할 수 있다.
