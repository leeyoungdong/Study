- 온라인 데이터베이스 백업은 크게 논리적 백업과 물리적 백업의 두 가지 범주로 분류할 수 있음.
  - 둘 다 장점과 단점이 있지만 논리적 백업의 한 가지 단점은 시간이 많이 걸릴 수 있다는 것.
  - 특히, 대용량 데이터베이스를 백업하는 데 시간이 오래 걸릴 수 있으며, 백업 데이터에서 데이터베이스를 복원하는 데는 더 오랜 시간이 걸릴 수 있음.
  - 반면, 물리적 백업은 훨씬 더 빠르게 만들고 복원할 수 있으므로 실제 시스템에서 매우 중요하고 유용한 기능.

- PostgreSQL에서는 버전 8.0부터 온라인 물리적 전체 백업이 가능했음.
  - 실행 중인 전체 데이터베이스 클러스터(예: 물리적 백업 데이터)의 스냅샷을 기본 백업 이라고 함.

- 버전 8.0부터 제공되는 PITR(Point-in-Time Recovery)은 지속적인 아카이브 기능 으로 생성된 기본 백업 과 아카이브 로그를 이용하여 데이터베이스 클러스터를 특정 시점으로 복원하는 기능
  - 예를 들어, 중대한 실수(예: 모든 테이블 자르기)를 한 경우 이 기능을 사용하면 실수가 발생하기 직전의 시점으로 데이터베이스를 복원할 수 있음.
