# **LOGICAL STRUCTURE OF DATABASE CLUSTER**

- 데이터베이스 클러스터 는 PostgreSQL 서버에서 관리하는 데이터베이스 모음
- PostgreSQL에서 '데이터베이스 클러스터'라는 용어는 '데이터베이스 서버 그룹'을 의미 하지 않음
- PostgreSQL 서버는 단일 호스트에서 실행되며 단일 데이터베이스 클러스터를 관리

<br>

- 데이터베이스 는 데이터베이스 개체의 모음
- 관계형 데이터베이스 이론에서 데이터베이스 개체는 데이터를 저장하거나 참조하는 데 사용되는 데이터 구조입니다
- (힙) 테이블 이 대표적인 예이며, 인덱스, 시퀀스, 뷰, 함수 등 기타 여러 가지가 있음
- PostgreSQL에서는 데이터베이스 자체도 데이터베이스 개체이며 논리적으로 서로 분리되어 있음
- 다른 모든 데이터베이스 개체(예: 테이블, 인덱스 등)는 해당 데이터베이스에 속함

<br>

**데이터베이스 클러스터의 논리적 구조**

- PostgreSQL의 모든 데이터베이스 객체는 부호 없는 4바이트 정수인 각 객체 식별자(OID) 에 의해 내부적으로 관리
- 데이터베이스 개체와 해당 OID 간의 관계는 개체 유형에 따라 적절한 시스템 카탈로그에 저장
- 예를 들어 데이터베이스와 힙 테이블의 OID는 각각 pg_database 와 pg_class 에 저장

**다음과 같은 쿼리를 실행하여 알고 싶은 OID를 찾을 수 있**



```
sampledb=# SELECT datname, oid FROM pg_database WHERE datname = 'sampledb';
 datname  |  oid
----------+-------
 sampledb | 16384
(1 row)

sampledb=# SELECT relname, oid FROM pg_class WHERE relname = 'sampletbl';
  relname  |  oid
-----------+-------
 sampletbl | 18740
(1 row)
```
