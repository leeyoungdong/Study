- 스트리밍 복제는 특별한 하드웨어가 필요하지 않은 두 가지 일반적인 오류 감지 절차를 사용.

1. 대기 서버 프로세스의 실패 감지:
  - Walsender와 Walreceiver 간의 연결 끊김이 감지되면 기본 서버는 즉시 대기 서버 또는 Walreceiver 프로세스에 결함이 있다고 판단.
  - 낮은 수준의 네트워크 기능이 월리시버의 소켓 인터페이스 쓰기 또는 읽기에 실패하여 오류를 반환하면 기본 서버도 즉시 실패를 판단.

2. 하드웨어 및 네트워크 오류 감지:
  - Walreceiver가 wal_sender_timeout 매개변수에 설정된 시간 (기본값 60초) 내에 아무 것도 반환하지 않으면 기본 서버는 대기 서버에 결함이 있다고 판단.
  - 위에서 설명한 오류와 달리 대기 서버가 일부 오류로 인해 더 이상 응답을 보낼 수 없는 경우에도 기본 서버에서 대기 서버의 소멸을 확인하는 데 최대 wal_sender_timeout 초의 일정 시간이 걸림.[대기 서버의 하드웨어 장애, 네트워크 장애 등]


<br>

- 장애 유형에 따라 일반적으로 장애 발생 직후 감지될 수 있음.
  - 그러나 장애 발생과 감지 사이에는 시간차가 있을 수 있음.
  - 특히 동기식 대기 서버에서 후자 유형의 오류가 발생하면 여러 잠재적 대기 서버가 작동 중이더라도 대기 오류가 감지될 때까지 기본 서버의 모든 트랜잭션 처리가 중지됨.
