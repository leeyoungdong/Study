- 동기식 스트리밍 복제는 버전 9.1에서 구현됨.
  - 이는 단일-마스터-다중 슬레이브 유형의 복제.
  - 여기서 "마스터"와 "슬레이브"라는 용어는 일반적으로 각각 기본 및 대기를 나타냄 .

- 이 기본 복제 기능은 기본 서버가 대기 서버에 WAL(Write-Ahead Log) 데이터를 지속적으로 전송한 후 수신된 데이터를 즉시 재생하는 일반적인 복제 기술인 로그 전달을 기반

**스트리밍 복제가 작동하는 방식**
- 스트리밍 복제가 시작되는 방법.
- 기본 서버와 대기 서버 간에 데이터가 전송되는 방식입.
- 기본 서버가 여러 대기 서버를 관리하는 방법.
- 기본 서버가 대기 서버의 장애를 감지하는 방법.

**Info**
- 비동기 복제 전용이었던 첫 번째 복제 기능은 버전 9.0에서 구현되었지만 버전 9.1에서는 동기 복제를 위한 새로운 구현(현재 사용 중)으로 대체
