- 트랜잭션 로그는 시스템 장애가 발생하더라도 데이터가 손실되지 않도록 보장하므로 데이터베이스의 필수적인 부분.
- 이는 데이터베이스 시스템의 모든 변경 사항 및 작업에 대한 기록 로그.
- 이렇게 하면 정전이나 서버 충돌과 같은 오류로 인해 데이터가 손실되지 않음.

<br>

- 로그에는 이미 실행된 각 트랜잭션에 대한 충분한 정보가 포함되어 있으므로 데이터베이스 서버는 서버 충돌 시 트랜잭션 로그의 변경 사항 및 작업을 재생하여 데이터베이스 클러스터를 복구할 수 있음.

- 컴퓨터 과학 분야에서 WAL은 Write-Ahead Logging 의 약어로 , 트랜잭션 로그에 변경 사항과 작업을 모두 기록해야 하는 프로토콜 또는 규칙.
  - 그러나 PostgreSQL에서 WAL은 Write Ahead Log 의 약어이기도 함 .
  - PostgreSQL에서 WAL이라는 용어는 트랜잭션 로그와 같은 의미로 사용되며, 트랜잭션 로그(WAL)에 작업을 기록하기 위해 구현된 메커니즘을 나타내기도 함.
  - 혼란스러울 수 있지만 이 문서에서는 PostgreSQL 정의를 채택함.

- WAL 메커니즘은 서버 충돌의 영향을 완화하기 위해 버전 7.1에서 처음 구현되었음.
- 또한 10 장 과 11 장에서 각각 설명하는 PITR(Point-in-Time Recovery) 및 SR(스트리밍 복제)의 구현도 가능해졌음.

- PostgreSQL을 사용한 시스템 통합 및 관리에는 WAL 메커니즘을 이해하는 것이 필수적이지만 이 메커니즘의 복잡성으로 인해 설명을 간략하게 요약하는 것은 불가능함.
  
**PostgreSQL의 WAL에 대한 전체 설명은 다음과 같음.**

- WAL(트랜잭션 로그)의 논리적 및 물리적 구조.
- WAL 데이터의 내부 레이아웃.
- WAL 데이터 작성.
- WAL 작성자 프로세스.
- 체크포인트 처리.
- 데이터베이스 복구 처리 중.
- WAL 세그먼트 파일 관리.
- 지속적인 아카이빙.
