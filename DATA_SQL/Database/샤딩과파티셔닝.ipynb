# **샤딩과 파티셔닝의 차이점?**

둘 다 데이터를 여러 시스템에 배포하고 스케일 아웃 데이터베이스 아키텍처를 구축하려고 합니다.

**파티셔닝**
* 데이터 항목(레코드) 세트_각 레코드에는 키가 존재.
* 키를 사용하여 이러한 레코드를 여러 다른 단위로 분할(분배) 할 수 있음.
* Oracle RDBMS에 대해 이야기할 때 테이블 파티셔닝은 오랫동안 사용. 
* 파티션 키를 사용하여 테이블의 레코드를 2개 이상의 파티션으로 세분화합니다. 
* 파티션은 여전히 ​​동일한 DB 인스턴스의 제어를 받음. 
* 다른 피어 파티션 및 다른 파티션되지 않은 테이블과 동일한 CPU, 메모리, I/O, 스토리지 리소스를 공유.

* Oracle은 해시, 범위 및 목록 기반 파티셔닝을 지원
* 다른 배포 체계의 목적은 레코드에 대한 키가 주어지면 그것이 속할 대상 파티션을 결정합니다.

**샤딩**

* 서로 다른 물리적 머신/노드에 걸친 파티셔닝을 샤딩
* 각 파티션은 완전히 다른 물리적 시스템 에 있으므로 다른 스키마 에 있고 별도의 데이터베이스 인스턴스 의 제어를 받음
* 여러 시스템에 데이터를 배포하는 접근 방식은 해시 및 범위입니다.

* 샤딩은 서로 다른 여러 서버에 데이터를 분산하는 방법입니다 
* 샤딩을 통해 수평적 확장성을 달성

**차이점**
* 샤딩은 여러 다른 시스템에 데이터를 분산 또는 분할하는 반면 파티셔닝은 동일한 시스템에 데이터를 분산하는 것
