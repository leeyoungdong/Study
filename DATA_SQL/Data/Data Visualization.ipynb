# **Data Visualization**

사람들이 맥락과 의미를 이해하는데 도움이 되는 데이터의 그래픽적인 표현. Interactive Data Visualization를 통해 회사는 Drill Down하여 셉 정보를 탐색하고, 패턴과 이상값을 식별 / 처리 /제외 데이터를 변경할 수 있음

* Interactive Data Visualization: 데이터의 그래픽 플로팅의 요소를 직접 조작 및 탐색하기 위한것(소프트웨어 사용 등등)
* Drill Down: 가장 요약된 레벨로부터 가장 상세한 레벨까지 차원의 계층에 따라 분석에 필요한 요약 수준을 바꿀 수 있는 기능

데이터 시각화는 분석을 이해하고, 전달하는 첫번째 단계일 경우가 많음. 수치가 아닌 그래픽으로 데이터를 시각화하면 더쉽게 추세 등을 확인 할 수 있으며 인사이트 도출로 이어짐.

# **데이터 시각화 비즈니스 기능**

**시각적 분석:**

시각적 분석 대시보를 사용하면, 가치가 높은 통찰력을 쉽고 빠르게 생성하고 제공할 수있음. 대시보드를 사용하면 조직이 통찰력을 사용해 더 나은 의사결정을 내릴 수 있음

ex. 스마트 데이터 검색, 데이터 랭글링, 지리공간 분석, 원클릭 예측분석

# **데이터 시각화의 이점**

**빠른 인사이트 도출**

모든 사람이 데이터를 빠르고 쉽게 인사이트로 변환할 수 있도록 지원. 데이터 시각화를 통해 추세를 발견, 이상값을 빠르게 식별할 수 있음. 비즈니스가 어떻게 수행되고 있으며 어떤위험히 발생할 것인지 이해하는 데 도움이 됨.

**똑똑하고 빠른 의사결정**

조직은 시각화를 통해 빠른 데이터 기반 의사결정을 내릴 수 있음. 실시간 데이터의 정보를 이해하고 다른사람들과 협력해 통창력을 촉발, 데이터의 패턴을 발견함으로 빠른 결정을 내리기 위해 직관을 버리고 논리적인 의사결정 가능

**고급 분석으로 역량강화**

공간, 클릭 예측 및 실시간 데이터 분석으로 더깊은 통찰력을 개발, 숨겨진 패턴을 발견, 높은 가치의 비즈니스 기회에 조치를 취할 수 있음

**제품 혹은 서비스의 가치향상**

데이터는 엄청난 가치를 지니고있지만, 대부분의 사용자들은 전문 BI도구를 배울 자원을 제공할 수없음. 앱과 비즈니스 프로세스에 분석기능을 내장해 사용자에게 데이터를 편리하고 직관적이게 만듬

**맞춤형 분석 요청에 소요되는 시간 단축**

기술팀이 보고서 OR 시각화에 대한 지정코드를 작성하고 반복 / 시간소요적인 작업에서 벗어나, 최종 사용자가 스스로 데이터를 시각화하고답을 얻을수 있도록 함

**데이터 시각화에 대한 최상의 접근 방식**
데이터 시각화에 많은 투자를 하는 회사의 경영진은 투자 결과를 기대. 그러나 생성된 시각화는 가치가 없거나 조직에 유용하지 않은 경우가 많음. 가치 있는 데이터 시각화를 생성하려면 설계, 데이터 랭글러 및 토픽 전문가가 필요함.  의미 있는 방식으로 데이터를 제공하기 위해 숙련된 팀을 구성해야 함. 사람들을 탁월한 업무 수행에 기여하는 기술과 도구와 연결함으로써의 최상의 결과물을 기대

**대화형 데이터 시각화의 영향**
대화형 데이터 시각화를 통해 사용자는 귀중한 정보를 자주 얻을 수 있습니다. 이는 의사 결정자가 시간에 민감한 상황에서 직감과 추측을 사용하는 위험을 피하는 데 도움이 됩니다. 대화형 데이터 시각화 기능을 통해 사용자는 더 나은 결정을 내리는 통찰력을 생성.
