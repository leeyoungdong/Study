# **Data Catalog란?**

데이터 카탈로그는 조직 내에서 구성된 데이터 자산 인벤토리, 메타데이터를 사용하여 조직의 데이터 관리 작업을 지원함. 또한 데이터 전문가가 메타데이터를 수집, 구성, 액세스 및 강화하여 데이터 검색 및 거버넌스를 지원하는 데 도움이 된다.

image.png

**데이터 카탈로그의 필요 이유**

각 조직은 점차 비즈니스 데이터가 증가하며, 점점 증가하는 대규모 데이터 애셋을 다루는데 효율적으로 활용하기위해필요하며 아래와 같은 문제와 직면한다.

* 데이터 검색 및 액세스에 시간과 노력 낭비
* 데이터 늪으로 변하는 데이터 레이크
* 일반적인 비즈니스 어휘 없음
* 이해하기 어려운 "다크 데이터"의 구조와 다양성
* 출처, 품질, 신뢰성을 평가하기 어려움
* 부족하거나 누락된 지식을 확보할 방법이 없음
* 지식과 데이터 자산을 재사용하기 어려움
* 수동적이고 임시적인 데이터 준비 노력

적절한 도구가 없으면 문제가 데이터를 효율적으로 사용하는 데 주요 장애물이 된다.


**데이터 카탈로그의 핵심 요소**


* 커넥터 및 큐레이션 도구

  데이터 카탈로그는 데이터에 대한 단일 신뢰 장소 역할을 한다. 커넥터는 데이터베이스의 물리적 데이터 세트를 매핑한다. 따라서 데이터 카탈로그를 강화하기 위해 다양한 커넥터를 보유하는 것이 중요. Salesforce, SQL 쿼리, 비즈니스 인텔리전스, 데이터 통합 도구와 같은 여러 소스에서 메타데이터를 수집할 수 있으므로 이 데이터도 선별하는 것이 중요. 검증 및 인증은 데이터 카탈로그의 효율성을 높이고 데이터 거버넌스를 지속 가능한 프로세스로 만드는 중요한 프로세스다.

* 효율적인 검색 옵션

  검색은 데이터 카탈로그의 기본 구성 요소. 사용자가 데이터 사이언스, 분석 또는 데이터 엔지니어링 관련 데이터 세트를 빠르게 찾을 수 있도록 유연한 검색 및 필터링 옵션이 있어야함. 데이터 자산의 기술적 계층을 기반으로 메타데이터를 찾기도함.

* 계보 또는 수명 주기 추적

  Lineage는 본 데이터의 수명 주기를 엿볼 수 있는 기회를 제공한다. 불일치가 있는 경우 사용자는 데이터 카탈로그를 사용하여 계보를 쉽게 추적하여 문제를 찾고 수정할 수 있다. 또한 조직의 다양한 소스와 데이터 유형 간의 차이점을 이해하는 데 도움이 됨.

* 범용 용어 및 데이터 사전

  조직의 데이터는 가치의 상당 부분을 차지하므로 모든 잠재적 이해 관계자가 액세스하고 쉽게 이해할 수 있어야 함. 일반적으로 데이터 카탈로그는 데이터 사전과 용어집으로 구성된다. 데이터 사전은 의미, 다른 데이터와의 관계, 출처, 용도 및 형식을 포함하여 카탈로그의 데이터에 대한 모든 메타데이터(일반적으로 테이블에 저장됨)의 모음. 용어집을 통해 조직 구성원은 카탈로그에 사용된 비즈니스 용어를 식별하고 회사 전체에서 동일한 방식으로 사용할 수 있어야 함.

* 프로파일링

  데이터 프로파일링은 데이터의 완전성, 정확성, 일관성 및 적시성을 평가하는 프로세스임. 데이터 프로파일링은 비즈니스 문제를 해결하기 위한 데이터의 유용성을 결정, 이는 여러 데이터 원본에서 데이터를 수집할 때 데이터 풀을 유지 관리하는 데 중요.

* 다양한 소스에서 메타데이터를 수집.

  데이터 카탈로그가 객체 스토리지, 자율 운영 데이터베이스, 온프레미스 시스템 등을 포함한 연결된 다양한 데이터 자산에서 기술 메타데이터를 수집할 수 있어야함.


**데이터 카탈로그로 구조화된 데이터에 대한 장점**

* 데이터의 품질, 구조, 사용 및 통계에 대한 정보를 포함하여 데이터 저장소 형성
* 실제 데이터와 함께 메타데이터에 액세스하면서 데이터에 대해 원격으로 협업가능.
* 자동으로 자주 업데이트하여 데이터 영역 전체에서 데이터가 정확하고 일관성이 있는지 확인 가능.
* 데이터 계보에 액세스하고 소스, 수정 및 데이터 액세스와 같은 정보 보기
* 안전한 방식으로 이해 관계자와 데이터 자산 공유 등...

**데이터 카탈로그 제품과 서비스 예시**

* AZURE 데이터 카탈로그와 AWS 글루(Glue)

  == 공용 클라우드 플랫폼에 내장된 데이터 카탈로그 서비스
* 인포매티카 엔터프라이즈 데이터 카탈로그(Informatica Enterprise Data Catalog), 탈렌드 데이터 카탈로그(Talend Data Catalog), SAP 데이터 허브(SAP Data Hub), IBM 인포스피어 인포메이션 거버넌스 카탈로그(IBM Infosphere Information Governance Catalog)

  == 데이터 통합 플랫폼의 데이터 카탈로그 기능
* 클라우데라 데이터 플랫폼, 인포웍스 데이터파운드리 등

  == 일부 데이터 카탈로그는 빅데이터 플랫폼과 하이브리드 클라우드를 위한 제품. 데이터 운영과 오케스트레이션을 지원한다.
* 유니파이 데이터 카탈로그(Unifi Data Catalog), 앨레이션 데이터 카탈로그(Alation Data Catalog), 콜리브라 카탈로그(Collibra Catalog), 워터라인 데이터(Waterline Data), IBM 왓슨 날리지 카탈로그(IBM Watson Knowledge Catalog)

  == 머신러닝 기능이 있는 독립형 플랫폼
* 스티보 시스템즈(Stibo Systems), 렐티오(Reltio)

  == 마스터 데이터 관리 툴과, 암  데이터(Arm Treasure Data)와 같은 고객 데이터 플랫폼의의 데이터 카탈로그 역할
