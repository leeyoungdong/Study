# **데이터 메시란?**

분산형 네트워크를 기반으로 하는 플랫폼 아키텍처. Data Warehouse나 Data Lake에서 중앙집중적으로 관리되었던 분석 데이터들을 탈중앙화하여 관리를 하는 개념. 데이터 소유권을 분산하고 도메인별 팀이 데이터를 독립적으로 관리, 데이터를 제품으로 취급, 데이터에 액세스하거나 관리하기 위해 중앙 집중식 시스템의 허가 없이 다른 최종 사용자에게 소유권을 분배함.

# **장점**

* 확장성:데이터 메시는 중앙 기관에서 액세스를 요청할 필요가 없는 기능 소유자의 손에 데이터 제어를 제공함, 이를 통해 조직은 운영 비용을 줄이고 작업 중복을 제거하며 연구 일정을 단축할 수 있다. 이러한 제어를 통해 조직은 더 짧은 기간에 제품과 서비스를 테스트하고 중앙 기관의 간섭을 거의 받지 않고 빠르게 결과를 얻을 수 있음.

* 구현시간: 데이터는 메쉬에서 제품으로 취급되기 때문에 기능 소유자는 높은 수준의 자율성을 갖는다. 중앙 위치에서 데이터를 보내고 받는 시간이 없어짐

* 독립: 데이터 메시의 분산형 접근 방식을 통해 기업은 벤더에 구애받지 않게 됩니다. 이를 통해 팀은 여러 시스템과 동시에 연결할 수 있다. 독립성은 디자인에 유연성을 제공하고 효과적임

* 투명도: 중앙 집중식 접근 방식은 전문가 팀을 사일로에 유지할 수 있다. 이것은 또한 조직의 투명성을 방해한다. 분산형 데이터 메시는 서로 다른 지정된 그룹에 걸쳐 데이터의 기능적 소유권을 허용한다. 이를 통해 전문가 팀이 데이터를 적절하게 제어하여 조직 내 투명성과 책임을 개선할 수 있다.

* 거버넌스 및 규정 준수: 분산된 소유권은 기업이 소스에서 시스템 보안을 제어할 수 있도록 도와준다. 전체 조직이 규정을 준수, 거버넌스 지침을 쉽게 따를 수 있도록 도와줍니다. 이러한 컴플라이언스를 통해 기업은 조직 내 데이터에 대한 액세스가 복잡하지 않은 상태에서 고품질에 맞게 데이터 전달을 최적화할 수 있다.

# **단점**

* 패러다임 변화: 데이터 메시의 주요 단점 중 하나는 조직 문화의 변화임. 시스템을 구현하려면 관련된 모든 당사자가 새로운 모델을 인식하고 지원하도록 조직 내에서 광범위한 훈련과 교육이 필요함. 최종 사용자에게 액세스 및 제어 권한을 부여하려면 시스템의 이점을 활용하는 방법을 이해해야 함. 데이터 메시는 최종 사용자에게 독립성을 제공하므로 자신의 행동이 조직에 어떤 영향을 미칠지 이해해야 함. 고도로 중앙 집중화된 운영 절차에 의존하는 전통적인 조직은 비효율성을 피하기 위해 변경 사항을 해결하고 변화를 관리해야 함.

* 높은 진입 비용:데이터 메시를 사용하려면 기존 인프라, 도구 및 소프트웨어를 새 모델에 맞게 조정해야 함. 따라서 조직에서 시스템에 데이터 메시 아키텍처를 채택할 때 기존 데이터를 메시에 통합하는 비용이 발생. 또한 조직은 새로운 메시 아키텍처를 지원하기 위한 인프라도 구축해야 함. 여기에는 다음에 대한 지원이 포함된다.데이터 통합, 정보 가상화, 거버넌스 및 규정 준수,데이터의 목록화등이 포함됨.

* 도메인 간 모니터링: 데이터 메시는 최종 사용자에게 데이터에 대한 효과적인 제어를 제공한다. 따라서 조직이 메시 아키텍처를 구현할 때 시스템을 모니터링할 수 있는지 확인하고, 이는 조직 내에서 효율성과 보안을 보장. 인증된 사용자만 데이터 및 정보에 액세스할 수 있고 이러한 데이터에 대한 액세스 권한이 중앙 위치에서 부여되도록 하려면 메시를 모니터링해야 함. 데이터 메시의 핵심은 데이터 및 액세스 분산에 있다.
